# American National Election Studies 2016

The data was downloaded from American National Election Studies [2016 Time Series Study](https://electionstudies.org/data-center/2016-time-series-study/).


## Study Overview
### Study Content Highlights:
In addition to content on electoral participation, voting behavior, and public opinion, the 2016 ANES Time Series Study contains questions in other areas such as media exposure, cognitive style, and values and predispositions. Several items first measured on the 2012 study were asked again including the ‘Big Five’ personality traits using the Ten Item Personality Inventory (TIPI), and skin tone observations made by interviewers in the face-to-face study. For the first time, ANES has collected supplemental data directly from respondents’ Facebook accounts. The Post-Election interview also included Module 5 from the Comparative Study of Electoral Systems (CSES).

### Study Design Highlights:
Data collection for the ANES 2016 Time Series Study began in early September and continued into January, 2017. Pre-election interviews were conducted with study respondents during the two months prior to the 2016 elections and were followed by post-election reinterviewing beginning November 9, 2016. As in 2012, face-to-face interviewing was complemented with data collection on the Internet. Data collection was conducted in the two modes independently, using separate samples but substantially identical questionnaires. Web-administered cases constituted a representative sample separate from the face-to-face.

## Rename Columns

These column names are too unfriendly to work with.  We can rename them

```python
interesting_cols = ['V161086', 'V161087', 'V160001', 'V160501', 'V161267', 'V161342', 'V165702', 'V165703', 
           'V165704', 'V165705', 'V161155']
df_clean = df[interesting_cols]
df_clean
```

```python
df_clean.rename(
    columns={
        'V161086': 'ft_dem',                       # * V161086: (PRE) Thermometer - Democratic Presidential Candidate
        'V161087': 'ft_rep',                       # * V161087: (PRE) Thermometer - Republican Presidential Candidate
        'V160001': 'caseid',                       # * V160001: Study Respondent Number: Year-Level CaseID
        'V160501': 'interview_mode',               # * V160501: Mode of Interview
        'V161267': 'age',                          # * V161267: (PRE) Respondent - Age
        'V161342': 'gender',                       # * V161342: Respondent - R self-identified gender
        'V165702': 'interview_start_date',         # * V165702: Interview Start Date
        'V165703': 'interview_start_time',         # * V165703: Interview Start Time
        'V165704': 'interview_end_date',           # * V165704: Interview End Date
        'V165705': 'interview_end_time',           # * V165705: Interview End Time
        'V161155': 'party_id'                      # * V161155: PRE: party ID: Does R think of self as Dem, Rep, Ind or what?
    }, 
inplace=True)

df_clean
```

## Remove FTF Data

FTF samples don't contain the start date or time of the interview

```python
df_clean = df_clean[df_clean["interview_mode"] == "2. Web"]
df_clean
```

### Is there anyone without a Feeling Thermometer?

```python
df_clean = df_clean[ ~df_clean["ft_dem"].str.startswith('-99', na=False) ]  # Refused
df_clean = df_clean[ ~df_clean["ft_dem"].str.startswith('-88', na=False) ]  # Don't Know
```

```python
df_clean = df_clean[ ~df_clean["ft_rep"].str.startswith('-99', na=False) ]
df_clean = df_clean[ ~df_clean["ft_rep"].str.startswith('-88', na=False) ]
```

## Proper column types
Fix the column types so that they are interpreted properly

```python
df_clean["ft_dem"] = pd.to_numeric(df_clean["ft_dem"])
df_clean["ft_rep"] = pd.to_numeric(df_clean["ft_rep"])
```

```python
df_clean["interview_start_date"] = pd.to_datetime(df_clean["interview_start_date"])
df_clean["interview_end_date"] = pd.to_datetime(df_clean["interview_end_date"])
```

```python
df_clean["interview_start_time"] = pd.to_datetime(df_clean["interview_start_time"], format="%H%M%S")
df_clean["interview_end_time"] = pd.to_datetime(df_clean["interview_end_time"], format="%H%M%S")
```

## Any interviews that crossed days?

```python
n_multidaysurveys = len(df_clean[df_clean["interview_start_date"] != df_clean["interview_end_date"]])
n_singledaysurveys = len(df_clean[df_clean["interview_start_date"] == df_clean["interview_end_date"]])



print(f"{n_multidaysurveys=}; {n_singledaysurveys=}")
```

## Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

demblue = "#0015BC"
repred = "#FF0000"
demgrey = "#9EA4BF"
repgrey = "#BF9EA2"
```
