## October Surprise 2016 

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_2016 = pd.read_csv("2016_clean.csv", parse_dates = ['interview_start_date', 'interview_end_date'])
```

```python
df_2016
```

```python
groupped_df = df_2016.groupby('interview_start_date', as_index=False)
```

```python
mean_ft_dem = df_2016.groupby('interview_start_date', as_index=False)['ft_dem'].mean()
mean_ft_dem = mean_ft_dem.set_index("interview_start_date")
mean_ft_dem
```

```python
mean_ft_rep = df_2016.groupby('interview_start_date', as_index=False)['ft_rep'].mean()
mean_ft_rep = mean_ft_rep.set_index("interview_start_date")
mean_ft_rep
```

```python
demblue = "#0015BC"
repred = "#FF0000"
demgrey = "#9EA4BF"
repgrey = "#BF9EA2"
```

```python
fig, ax = plt.subplots(figsize=(16, 8))

sns.lineplot(
    data = mean_ft_dem, 
    ax = ax,
    x = "interview_start_date", 
    y = "ft_dem",
    color = demblue
)
sns.lineplot(
    data = mean_ft_rep, 
    ax = ax,
    x = "interview_start_date", 
    y = "ft_rep",
    color = repred
)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel("Interview Start Date", fontsize=16)
ax.set_ylabel("Feeling Thermometer Temperature", fontsize=16)
ax.set_title("2016 ANES Survey Feeling Thermometer", fontsize=22)
```

```python
import datetime

start_date = datetime.datetime(2016, 10, 25)
end_date = datetime.datetime(2016, 11, 1)

print(f"{start_date=}; {end_date=}")
```

```python
df_surprise = df_2016[ df_2016["interview_start_date"] >= start_date ]
```

```python
df_surprise = df_surprise[ df_surprise["interview_start_date"] <= end_date ]
```

```python
groupped_df = df_surprise.groupby('interview_start_date', as_index=False)
```

```python
mean_ft_dem_surprise = df_surprise.groupby('interview_start_date', as_index=False)['ft_dem'].mean()
mean_ft_dem_surprise = mean_ft_dem_surprise.set_index("interview_start_date")
mean_ft_dem_surprise
```

```python
mean_ft_rep_surprise = df_surprise.groupby('interview_start_date', as_index=False)['ft_rep'].mean()
mean_ft_rep_surprise = mean_ft_rep_surprise.set_index("interview_start_date")
mean_ft_rep_surprise
```

```python
fig, ax = plt.subplots(figsize=(16, 8))

sns.lineplot(
    data = mean_ft_dem_surprise, 
    ax = ax,
    x = "interview_start_date", 
    y = "ft_dem",
    color = demblue
)
sns.lineplot(
    data = mean_ft_rep_surprise, 
    ax = ax,
    x = "interview_start_date", 
    y = "ft_rep",
    color = repred
)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel("Interview Start Date", fontsize=16)
ax.set_ylabel("Feeling Thermometer Temperature", fontsize=16)
ax.set_title("2016 ANES Survey Around October Surprise", fontsize=22)
```

```python
df_twitter_rep = pd.read_csv('data/trump_twts_2016_classified_senti+poli.csv', parse_dates=['time'])
#df_twitter_rep.head()
```

```python
df_twitter_dem = pd.read_csv('data/clinton_twts_2016_classified_senti+poli.csv',  parse_dates=['time'])
#df_twitter_dem.head()
```

```python
dfClinton = pd.read_csv("data/2016_twitter_dem_mean_sentiment.csv", parse_dates=['time'])
dfTrump = pd.read_csv("data/2016_twitter_rep_mean_sentiment.csv", parse_dates=['time'])
```

```python
df_twitter_rep_surprise = dfTrump[ (dfTrump["time"].dt.date >=  start_date.date()) & (dfTrump["time"].dt.date <=  end_date.date()) ]
df_twitter_rep_surprise
```

```python
df_twitter_dem_surprise = dfClinton[ (dfClinton["time"].dt.date >=  start_date.date()) & (dfClinton["time"].dt.date <=  end_date.date()) ]
df_twitter_dem_surprise
```

```python
fig, ax = plt.subplots(figsize=(16, 8))

sns.lineplot(
    data = df_twitter_dem_surprise, 
    ax = ax,
    x = "time", 
    y = "sentiment_label",
    color = demblue
)
sns.lineplot(
    data = df_twitter_rep_surprise, 
    ax = ax,
    x = "time", 
    y = "sentiment_label",
    color = repred
)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel("Date", fontsize=16)
ax.set_ylabel("Sentiment Score", fontsize=16)
ax.set_title("2016 Twitter Sentiment Around October Surprise", fontsize=22)
```

```python
fig, ax = plt.subplots(figsize=(16, 8))

sns.lineplot(
    data = df_twitter_dem_surprise, 
    ax = ax,
    x = "time", 
    y = "sentiment_label",
    color = demblue,
    linewidth = 5
)
sns.lineplot(
    data = df_twitter_rep_surprise, 
    ax = ax,
    x = "time", 
    y = "sentiment_label",
    color = repred,
    linewidth = 5
)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel("Date", fontsize=16)
ax.set_ylabel("Twitter Sentiment Score", fontsize=16)
ax.set_title("2016 Sentiment Around October Surprise", fontsize=22)


plt.legend(
    labels = ["Mean Twitter Sentiment (Dem)","Mean Twitter Sentiment (Rep)"], 
    loc = 3
)

ax2 = plt.twinx()

sns.lineplot(
    data = mean_ft_dem_surprise, 
    ax = ax2,
    x = "interview_start_date", 
    y = "ft_dem",
    color = demgrey,
    linewidth = 5
)
sns.lineplot(
    data = mean_ft_rep_surprise, 
    ax = ax2,
    x = "interview_start_date", 
    y = "ft_rep",
    color = repgrey,
    linewidth = 5
)

ax2.spines['left'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.spines['bottom'].set_visible(False)

ax2.set_ylabel("Feeling Thermometer Temperature (Rescaled)", fontsize=16)

plt.legend(
    labels = ["Mean Feeling Temperature (Dem)","Mean Feeling Temperature (Rep)"], 
    loc = 4
)


plt.axvline(
    pd.Timestamp('2016-10-28'),
    color = "k",
    linestyle = "--"
)
```
