# American National Election Studies 1948-2020

The data was downloaded from American National Election Studies [Time Series Cumulative Data File (1948-2020)](https://electionstudies.org/data-center/anes-time-series-cumulative-data-file/).


## Study Overview
Questions that have been asked in three or more ANES Time Series studies may be included in the ANES Time Series Cumulative Data File. Data from other ANES studies, such as the 1984 Continuous Monitoring Study, the 1988 Super Tuesday Study, or the 1988-1992 Senate Election Study, are never included.

## Interesting Columns

* VCF0424: Thermometer - Democratic Presidential Candidate
* VCF0426: Thermometer - Republican Presidential Candidate

* VCF0403: Affect (Likes minus Dislikes) - Democratic Presidential Candidate
* VCF0407: Affect (Likes minus Dislikes) - Republican Presidential Candidate

* VCF0004: Year of Study
* VCF0006: Study Respondent Number: Year-Level CaseID
* VCF0009z: Weight (???)
* VCF0017: Mode of Interview
* VCF0101: Respondent - Age
* VCF0104: Respondent - Gender
* VCF0803: Liberal-Conservative Scale
* VCF0301: PARTY IDENTIFICATION OF RESPONDENT- 7-POINT SCALE
* VCF0302: PARTY IDENTIFICATION OF RESPONDENT- INITIAL PARTY ID RESPONSE
* VCF0303: PARTY IDENTIFICATION OF RESPONDENT- SUMMARY 3- CATEGORY

## Rename Columns

These column names are too unfriendly to work with.  We can rename them

```python
interesting_cols = ['VCF0424', 'VCF0426', 'VCF0403', 'VCF0407', 'VCF0004', 'VCF0006', 'VCF0009z', 'VCF0017', 
           'VCF0101', 'VCF0104', 'VCF0803', 'VCF0301', 'VCF0302', 'VCF0303']
df_clean = df[interesting_cols]
df_clean
```

```python
df_clean.rename(
    columns={
        'VCF0424': 'ft_dem',          # VCF0424: Thermometer - Democratic Presidential Candidate
        'VCF0426': 'ft_rep',          # VCF0426: Thermometer - Republican Presidential Candidate
        'VCF0403': 'affect_dem',      # VCF0403: Affect (Likes minus Dislikes) - Democratic Presidential Candidate
        'VCF0407': 'affect_rep',      # VCF0407: Affect (Likes minus Dislikes) - Republican Presidential Candidate
        'VCF0004': 'year',            # VCF0004: Year of Study
        'VCF0006': 'caseid',          # VCF0006: Study Respondent Number: Year-Level CaseID
        'VCF0009z': 'weight',         # VCF0009z: Weight 
        'VCF0017': 'interview_mode',  # VCF0017: Mode of Interview
        'VCF0101': 'age',             # VCF0101: Respondent - Age
        'VCF0104': 'gender',          # VCF0104: Respondent - Gender
        'VCF0803': 'lib_con_scale',   # VCF0803: Liberal-Conservative Scale
        'VCF0301': 'party_id_7',      # VCF0301: PARTY IDENTIFICATION OF RESPONDENT- 7-POINT SCALE
        'VCF0302': 'party_id_init',   # VCF0302: PARTY IDENTIFICATION OF RESPONDENT- INITIAL PARTY ID RESPONSE
        'VCF0303': 'party_id_3',      # VCF0303: PARTY IDENTIFICATION OF RESPONDENT- SUMMARY 3- CATEGORY
    }, 
inplace=True)

df_clean
```

## Remove Older Data

There was no election before the first tweet in 2006

```python
df_clean = df_clean[df_clean["year"] > 2006]
df_clean
```

## 2008 Election

Pre-election interviews 73 minutes in length were conducted September 2 through November 3, 2008. No interviewing was conducted on Election Day, November 4. Post-election interviews 91 minutes in length were administered November 5 through December 30, 2008. 

```python
df_2008 = df_clean[df_clean["year"] == 2008]
df_2008
```

```python
fig, ax = plt.subplots()
sns.histplot(
    data = df_2008,
    x = "ft_dem",
    ax=ax,
    palette=[demblue],
)
```

```python
fig, ax = plt.subplots()
sns.histplot(
    data = df_2008,
    x = "ft_rep",
    ax=ax,
    palette=[repred],
)
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
sns.kdeplot(
    data = df_2008,
    x = "ft_dem",
    ax = ax,
    color = demblue,
    label = 'Obama'
)
sns.kdeplot(
    data = df_2008,
    x = "ft_rep",
    ax = ax,
    color = repred,
    label = 'McCain'
)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel("Feeling Thermometer Temperature", fontsize=16)
ax.set_ylabel("Density of Respondents", fontsize=16)
ax.set_title("2008", fontsize=22)

plt.legend()
```

## 2012 Election


As with all Time Series studies conducted during years of presidential elections,
respondents were interviewed during the two months preceding the November
election (Pre-election interview) and then re-interviewed during the two months
following the election (Post-election interview). Face-to-face interviews were
conducted with one interview before the election and a second interview with the
same respondents after the election. Internet interviews were conducted with two
interviews before the election and two more interviews with the same respondents
after the election. Internet data collection was managed by GfK. In-person data
collection was managed by Abt SRBI.

```python
df_2012 = df_clean[df_clean["year"] == 2012]
df_2012
```

```python
fig, ax = plt.subplots()
sns.histplot(
    data = df_2012,
    x = "ft_dem",
    ax=ax,
    palette=[demblue],
)
```

```python
fig, ax = plt.subplots()
sns.histplot(
    data = df_2012,
    x = "ft_rep",
    ax=ax,
    palette=[repred],
)
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
sns.kdeplot(
    data = df_2012,
    x = "ft_dem",
    ax = ax,
    color = demblue,
    label = 'Obama'
)
sns.kdeplot(
    data = df_2012,
    x = "ft_rep",
    ax = ax,
    color = repred,
    label = 'Romney'
)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel("Feeling Thermometer Temperature", fontsize=16)
ax.set_ylabel("Density of Respondents", fontsize=16)
ax.set_title("2012", fontsize=22)

plt.legend()
```
